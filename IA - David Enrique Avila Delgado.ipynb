## Introducción

¿Qué es la Inteligencia Artificil?, ¿cómo ha ido evolucionando a partir del tiempo?, ¿qué podemos esperar de esta nueva ciencia tecnológica? Estas y otras preguntas frecuentes se han presentado en todos aquellos que en algún momento de su vida han escuchado hablar de la Inteligencia Artificial o bien estudiado acerca de esta misma, ya que si bien es cierto que es una de las ciencias modenras, la verdad es que no es tan nueva como muchas de las personas solemos imaginar. La inteligencia Artificial comenzó poco después de la Segunda Guerra Mundial, ya que fue en el año de 1943 con la prublicación del artículo creado por Warren McCullough y Walter Pitts llamado "A Logical Calculus of Ideas Immanent in Nervous Activity", donde estos científicos plantearon por primera vez un modelo para la creación de una red neuronal.

A lo largo de la historia coneptualizar lo qué es la Inteligencia Artificial ha sido muy complejo, ya que de cierta forma, para algunos autores es un enfoque el cual está centrado en el comportamiento humano que se considera que no debería ser una ciencia empírica, si no más bien debe incluir hipóteis y confirmaciones mediante expermientos, sin embargo, para otros autores, este es un enfoque está centrado en torno a la racionalidad el cual implica uan combinación de matemáticas e ingeniería en su más alto nivel.

A continuación, se presentan las definiciones de  de varios autores respecto a su enfoque.

En el año de 1950 se crea el primer ordenador de red neuronal, el cual es creado por dos alumnos de la universidad de Harvard: Marvin Minsky y Dean Edmons justo el mismo año en el que Alan Turign publicó que hoy en día conocemos como los valores de IA, llamado Test de Turing.

La prueba de Turing diseñada para proporcionar una definición operacional y satisfactoría de lo que hoy en día conocemos como inteligencia. La prueba consite en que el computador debe superar dicha prueba con la condición de que el evaluador humano no sea capaz de distinguir si las respuestas planteadas a la Inteligencia Artificil son generadas por una persona o un máquina, es en ese momento donde consideramos que realmente es una Inteligencia Artificial, para ello, el computador debe poseer las siguientes capacidades:
* **Procesamiento de lenguaje natural**. Permitirá comunicarse satisfactoriamente en cualquier idioma en el que sea programado.
* **Representación del conocmiento**. Deberá poder almacenar lo que se conoce o siente.
* **Razonamiento automático**. Deberá contar con la capacidad de contar de utilizar dicha información alamacenada para respodnder a todo tipo de preguntas y poder extraer así nuevas conclusiones.
* **Aprendizaje automático**. Debe poder adaptarse a las nuevas circunstancias para detectar y extrapolar patrones.

Esta prueba de Turing evitó deliberadamente toda aquella interacción física que sea directa entre el computador y el evaluador, ya que para medir dicha inteligencia sería innecesario simular a un persona física, sin embargo, tiempo después surgió lo que se le conoce como la Prueba Global de Turing, en la cual se incluye una señal de video que permite al evaluador valorar la capacidad de percepción del evauluado, lo que se refiere a que el evaluador muestre o bien pase objetos físicos a través de una ventana y la Inteligencia Artificial tenga la capacidad tanto de manipularlo como de identificarlo, para ello, dicho computador de contar con las siguientes características:
* Visión computacional para percibir los objetos
* Robótica para manipular y mover objetos


# Watson y su impacto en el mundo

## Introducción

"Cuando una máquina muestra la capacidad de ver, escuchar, comprender y aprender sobre el mundo al igual que nosotros se le "considera artificialmente inteligente".  

En la actualidad el tema de la Inteligencia Artificial es cada día más tangible para todas las personas, incluso aquellas que no se encuentran en una rama de las ciencias computacionales, esto debido al gran impacto que ha tendió ChatGPT, Bing y muchas otras herramientas al implementar un chatbot el cual funciona a través de la Inteligencia Artificial, sin embargo, la Inteligencia Artificial no es una ciencia de la última década.

La Inteligencia Artificial comenzó poco después de la Segunda Guerra Mundial y adopto el nombre en el año de 1965, como tal la IA en la actualidad cuenta con una gran variedad de subcampos que van desde áreas de propósito general, como son el aprendizaje y la percepción, a áreas que son más específicas como es el caso de la demostración de teoremas matemáticos, escritura de poesía y el diagnóstico de enfermedades. La IA se considera potencialmente relevante para cualquier ámbito de la actividad intelectual humana, ya que presenta un sentido genuinamente universal.

El término cognitivo hace referencia a la capacidad de procesar información, aprender, razonar, resolver problemas, memorizar y tomar decisiones, por lo tanto, el término "computación cognitiva" hace referencia a que las máquinas imitan estas funciones para hacerlas inteligentes. En los últimos años gracias a los avances tecnológicos es posible contar con entornos de fabricación inteligentes, lo cual es una mejora a la productividad, procesos productivos más óptimos y comunicaciones más ágiles y robustas para la toma de decisiones más eficientes.

Las ventajas que presenta la Inteligencia Artificial son muchas considerando el gran número de enfoques que esta tiene así como las vertientes de cada uno de estos enfoques, es por ello que el estudio de la Inteligencia Artificial se ha vuelto demandado en las últimas décadas. Una de las herramientas pioneras en el manejo de Inteligencia Artificial de procesamiento de información es Watson, creada por IBM anunciada al público un 30 de mayo de 2006, un proyecto que buscaba desarrollar una computadora capaz de entender y responder preguntas en lenguaje natural.
    
¿Qué es IBM Watson? ¿para qué sirve esta IA de IBM? ¿qué ha significado en la actualidad contar con herramientas como Watson? Pues bien, Watson se considera como una plataforma de Inteligencia Artificial la cual permite incorporar herramientas de IA a datos de información, sin importar el alojamiento de estos, la cual logra a través de diversas herramientas usar estos datos para mejorar la experiencia de usuario, optimizar, procesos y diseñar flujos de trabajo.  

En este ensayo nos enfocaremos en la forma en la que la Inteligencia Artificial nos sirve como una herramienta de búsqueda esto debido a su gran capacidad de procesar más información de la que podría procesar un humano en toda su vida, así como la forma en la que funciona Watson, sus inicios, cómo fue evolucionado  y su repercusión en el mundo actual, más específicamente en el área de la salud con Watson Health.
    
## Desarrollo    

IBM Watson surge como un proyecto de investigación que arrancó en 2005, el cual tenía el objetivo de comprender el lenguaje natural de forma fluida y contar con la capacidad de responder preguntas. Fue hasta el año 2011 cuando Watson fue presentado en un programa de televisión Jeopardy! con la finalidad de enfrentarse a los mejores concursantes de dicho programa, Watson sorprendió al mundo entero cuando se enfrentó con éxito a los dos mejores concursantes en la historia del programa y vencerlos con facilidad.

El lenguaje natural no es cosa sencilla, para nosotros como seres humanos no es sencillo comprenderlo a través del tiempo, es tal el grado de entendimiento que en realidad no nos damos cuenta de lo complejo que es el lenguaje natural y el entendimiento que este conlleva ya que en muchas ocasiones formular y comprender una oración es algo complejo, tal es el caso de la famosa cita de Groucho Marx, la cual dice, "Una mañana me desperté y le disparé a un elefante en pijama. ¿Cómo se metió a mi pijama? Nunca lo sabré". Se pone en reto el entendimiento de la máquina, ya que esta debe desglosar la oración en oraciones más sencillas, las cuales cumplan con la forma principal del enunciado, artículo, sujeto y predicado, una vez desglosada cada una de las oraciones, la máquina debe comprender el sentido que lleva la oración para de esta forma crear un entendimiento y analizar el lenguaje natural.

Watson utilizó lo que se le conoce como aprendizaje automático, es decir, aquella subcategoría de la Inteligencia Artificial por la cual los PC desarrollan el reconocimiento de patrones o la capacidad de aprender continuamente, con esto poder realizar predicciones basadas en datos, por lo cual se realizarán ajustes sin haber sido programadas con ese fin.
    
Watson participó en Jeopardy! a través de un enfoque que se divide en los siguiente:
* **Aprendizaje automático**. Utilizado por Watson para el reconocimiento de patrones.
* **Dividir la información**. Watson divide la información para que contiene para entregar resultados.
* **Búsqueda de respuestas**. Una vez dividida la información busca respuestas las cuales coincidan.
* **Compila y analiza probabilidades**. Esto lo hace a través de miles de algoritmos.
* **Genera niveles de confianza**. Una vez compilado, genera niveles de confianza y los ordena.
* **Genera una respuesta**. Genera una respuesta en segundos una vez terminado todo el proceso.

Una vez que Watson tuvo éxito el doctor Noam Slonim decidió llevarlo más allá del concurso de televisión Joepardy! tomó la decisión de llevarlo más allá y ponerlo a prueba en una prueba de lenguaje natural de gran demanda, es decir, en debates, Watson, no solo tenía límite de tiempo para responder, buscando entre su gran cantidad de información un punto el cual se encontraba a favor o en contra de determinado tema, sino que también debía tener la capacidad de escuchar lo que el otro ponente argumentaba de manera rápida, clara y eficaz. Watson debía tener la capacidad de analizar la información recibida por el otro ponente y a la vez buscar la manera de refutar los argumentos contarios, para todo esto resultaba un reto bastante grande, además de contar con la facilidad de realizar chistes en contra del ponente contrario, esto solo con la finalidad de asemejarse a los debates entre personas, los cuales suelen hacer este tipo de chistes.

Una vez que Watson fue programada para llevar a cabo este aprendizaje automático lo realizó a través de las siguientes fases:
* **Escanear millones de artículos**. Watson escaneaba una gran cantidad de artículos en búsqueda de aquellos que le presentaban puntos a favor respecto a su postura dentro del debate y puntos en contra respecto a la postura del otro ponente.
* **Procesamiento del lenguaje natural**. Watson debía procesar cada uno de los argumentos propios y contarios, con la finalidad de llevar a cabo un debate lo suficientemente convincente para el público en general.
* **Escuchar al oponente**. Watson escuchaba al oponente a la vez que este analizaba la información respecto a su gran cantidad de información con la que cuenta.
* **Entender la esencia**. Una vez analizado cada una de las palabras mencionadas por el oponente, Watson debía enfocarse únicamente en la esencia de la ponencia anterior para buscar argumentos a su favor.
* **Organización de argumentos**. Una vez procesados los millones de artículos de información, debía buscara la manera de argumentar sus puntos, de tal forma que refutaran los puntos propuestos por el ponente contrario.

Watson a lo largo del tiempo presentó muchas mejoras respecto a la forma en la que iba aprendiendo y desenvolviéndose a lo largo del debate, de tal manera que iba subiendo sus estadísticas de convencimiento hacia las personas, enfrentándose a una de las mejores debatientes que existe actualmente. Watson con el paso del tiempo y debido a su entrenamiento tuvo éxito en la prueba y logró construir su propia reputación.

Con los avances que presentó Watson en sus primeros retos, los desarrolladores se dieron cuenta de la gran cantidad de oportunidad que ofrece contar con plataformas de Inteligencia Artificial, es por ello que a partir de ese momento, IBM dedicó sus esfuerzos a crear aplicaciones y servicios con la ayuda de esta herramienta, a continuación, se presentarán varias de estas aplicaciones y servicios.

### Machine Learning
Con IBM Watson Machine Learning, nos permite en gran medida acelerar el despliegue de la Inteligencia Artificial y el machine learning, esto gracias a que Watson ayuda a simplificar y aprovechar a gran escala cualquier alojamiento en la nube, por lo que llevará a producción mucho más rápido y de forma más sencilla todos aquellos proyectos de código abierto.
Gracias a este servicio logramos desplegar modelos creados con IBM Watson Studio y herramientas de código abierto, así como entrenar los modelos de machine learning de forma dinámica, crear aplicaciones las cuales estén basadas en inteligencia artificial de forma más sencilla y agilizar la gestión gracias a su propia interfaz.

### Descubrimiento del Dato (Insights)
Watson trabaja con dos herramientas las cuales ayudan a anticipar variables utilizando sistemas de filtrado de información, las cuales son las siguientes:

#### Watson Discovery
Esta es una tecnología de búsqueda de Inteligencia Artificial y búsqueda empresarial la cual permite analizar todos aquellos datos los cuales nos ayudan a encontrar tendencias y respuestas específicas a ciertas preguntas, lo cual facilita la toma de decisiones y acelerar los procesos.
Watson Discovery utiliza machine learning para aprender y agilizar con el tiempo el análisis de cualquier tipo de datos, es decir, datos sin importar el formato en el cual se encuentren, así como clasificar la información por relevancia y ofrecer respuestas más acertadas a cada cliente con su integración de  chatbots.

#### IBM Watson Knowledge Catalog
Este es un catálogo de datos, los cuales está basado en IBM Cloud Pack for Data, la cual está integrada en una plataforma de Gobierno del Dato, el cual nos ayuda a encontrar, preparar, comprender y utilizar con facilidad la información que requerimos, además de que este catálogo nos ofrece la capacidad de proteger y controlar todos aquellos datos confidenciales de la empresa que requiera contar con este servicio.

### Procesamiento de lenguaje natural
Como pudimos observar en sus inicios, uno de los grandes problemas con los que se enfrentó esta Inteligencia Artificial fue comprender y analizar el lenguaje natural, sin embargo, una vez que este fue dominado por IBM Watson fue que ahora ofrecen el servicio llamado Watson Natural Language Understanding, la cual tiene la capacidad de obtener y clasificar todos aquellos datos importantes como palabras clave, sentimientos, emociones y roles semánticos, los cuales a partir de textos y conversaciones, está diseñada para comprender con total confianza el lenguaje humano.
Con Watson Knowledge Studio, se le podrá enseñar a comprender la terminología específica de la empresa o industria que cuente con este servicio, facilitando en gran medida el análisis, selección y clasificación de la información.
Con esta herramienta contamos con uno de los principales usos en la aplicación para chatbots, debido a su relevancia al momento de comprender interacciones humanas.

### Reconocimiento de voz
IBM Watson de igual forma cuenta con la capacidad de comprender el lenguaje escrito, es por ello, que a través de sus servicio "Watson Speech to Texto", la cual es una herramienta que tiene la capacidad de transformar de forma clara todo audio en texto, con la finalidad de agilizar la comprensión del contenido.
Watson cuenta con la capacidad de entender audios en 7 idiomas en tiempo real y de forma precisa, al igual que interpreta el tono y sentimiento de una  conversación, ya que al contar con funciones de machin learning y funciones como "Watson Language Translator" y "Watson Tone Analizer", su capacidad de comprensión se va adaptando a las necesidades cada vez más.

### Visión computacional
IBM trae una de las herramientas más importantes en el área de la Inteligencia Artificial, la cuál es la visión computacional, que permite una amplia gama de funcionalidades en la actualidad, tal como el manejo autónomo, reconocimiento fácil, entre muchas otras.
Con Watson Visual Recognition, IBM nos facilita la búsqueda, clasificación y etiquetado de contenido visual, ya que cuenta con algoritmos de deep learing para analizar imágenes de objetos, caras y otros contenidos.
Este tipo de funciones ayudan a ser más eficientes en procesos como auditorías visuales, análisis de incidentes para asegurados, comercio minorista, educación entre otras.

### IBM Watson aplicado a los negocios
Una de las aplicaciones más relevante en la actualidad con la que cuenta Watson, es aplicar su Inteligencia Artificial los negocios y áreas de la vida cotidiana de las personas, ya que es capaz de ayudar a asesores de inversión a recomendar los mejores planes de jubilación para los clientes que cuenten con Watson, además de ayudar a los call center a encontrar las respuestas más rápido o bien ayudar a los investigadores a encontrar patrones escondidos en graves volúmenes de datos.
Empresas reconocidas como Repsol, utilizan Watson para mejorar el acierto de búsquedas de petróleo, Honda para atendes a sus clientes, General Motors lo aplica para comprender mejor las preferencias de los consumidores y tomar decisiones de fabricación y marketing, Citigroup utiliza esté herramienta para recopilar información de sus clientes y obtener sugerencia para mejorar las interacciones, evaluar el riesgo e identificar oportunidades.

### Watson Health y su enfoque en la medicina
El uso de las computadoras y las Tecnologías de la Información dentro de los campos de la medicina, no es tema nuevo, al contrario, es un tema que se ha llevado a la práctica común, actividades como registro de historiales, detalles, enfermedades, entre muchas otras. 
Proveedores, líderes e investigadores de cuidado médico en la actualidad están utilizando tecnologías de Inteligencia Artificial para aprovechar millones de informes médicos, registros de pacientes, ensayos y revistas médicas con el propósito de encontrar insights de datos.

En el año 2015, IBM creó la unidad de negocio conocida como Watson Health, la cual radica en Boston y Nueva York, cuenta actualmente con 2,000 profesionales entre consultores, médicos, personal sanitario, desarrolladores e investigadores que trabajan para "desarrollar y acelerar la adopción de capacidades cognitivas que mejores la efectividad de la atención sanitaria.

En la actualidad IBM Watson puede integrarse con cualquier problema médico, lo cual supone un gran avance médico, así el usuario puede crear sus propias aplicaciones de IoT (Internet de las cosas) para cualquier enfermedad que este desee, ya que Watson no impone restricciones, sin embargo, ¿cómo es que la Inteligencia Artificial analiza una consulta? Pues bien, esto lo realiza con los siguientes pasos:
* Descompone la consulta mediante el análisis sintáctico y semántico en frases las cuales solo contengan información relevante. A través de algoritmos busca las parafrasees, elementes de tipo geográfico, establecer equivalencias de contenido, entre otras.
* La mayor preocupación es que la actividad se realiza en tiempo real, dando la posibilidad de tomar decisiones los cuales realizan procesos paralelos para construir posibles respuestas lo más grandes que sean posibles.

Funciones en las que se aplica Watson:
* **Consultor**. Un sistema que da soporte a la persona a partir de consultas de lenguaje natural.
* **Asesor**. Plantea una respuesta a partir de relacionar información relevante, tal es el caso de buscar fármacos que tengan sustancias químicas determinadas.
* **Asesor oncológicos**. Sistema que ayuda a la recomendación del tratamientos de escenarios que tengan las características del paciente, lo que permite mejorar la calidad de vida de los pacientes.
* **Asesor de control**. Es decir, en caso de que algún patrón de información se adecua o no al a hora de tomar decisiones.

#### Watson, Lector del genoma
La Inteligencia Artificial de IBM ha participado en varias ocasiones en estudios sobre medicina genómica, una fue en el Instituto Barrow y con el Centro del Genoma de Nueva York, en este último se probó un prototipo único de Watson, el cual tuvo como propósito específico el analizar genes con el objetivo de colaborar junto a los médicos a desarrollar tratamientos personalizados para pacientes que cuentan glioblastoma, el cuál es un cáncer muy agresivo que actualmente manta a 13,000 personas al año en Estados Unidos.

Par los especialistas, contar con herramientas necesarias para ofrecer a los enfermos un tratamiento basado en el análisis del ADN, fue donde intervino Watson, el cuál estableció correlaciones entre los datos de la secuencia genómica con toneladas de artículos médicos, informes e historiales clínicos, los cuales según los expertos este se duplica cada cinco años. Watson y el Centro del Genoma buscaban identificar patrones en la secuenciación y en los datos médicos que ayuden a los especialistas a hacer realidad la medicina genómica.

#### Watson y el Centro de Trastornos del Movimiento
Dentro del Documental de Discovery Channel llamado Inteligencia Artificial-IBM, nos presentan parte de las aplicaciones que tiene Watson, en este caso con el Centro de Trastornos del Movimiento, nos presentan el caso de Jonathan Rezek, el cuál es diagnosticado con Parkinson, el cual le impide desarrollarse en sus actividades diarias, sin embargo, con la colaboración con IBM Watson, se busca encontrar patrones de su enfermedad que lo ayuden a diagnosticar y tratarlo, además de que contar con una cantidad de millones de artículos les facilita al momento de encontrar fármacos con patrones similares que puedan tratar los síntomas del Parkinson.

# Teoría de las Inteligencias Múltiples

## Introducción
Desde la Antiguedad, el ser humano se ha envuelto en un mar de interrogantes, esto debido a que nuestra sociedad ha tratado de comprender desde sus inicios el mundo, lo que nos rodea y el cómo nos vemos afectados por este mismo, desde animismo o antropormorfismo a fenómenos naturales como lo pueden ser la lluvia, los truenos, etc. es por ello, que los seres humanos tendemos a la búsqueda del conocimiento, un claro ejemplo de esto es que desde los poemas escritos por Homero, ya se empezaba considerar el concepto de mente, un movimiento de personas conocidos como los protopsicólogos ya le otrogaban la capacidad a los seres humanos de algunas capacidades como lo son la memoria, pensamiento y entendimiento, otros filósofos como es el caso de Platón y aristótles también realizaron observaciones al campo de la inteligencia. Ya en la Edad Media fueron autores como Aquino y Buenaventura los que plantearon puntos de vista sobre la mente más próximos al aritotelismo y platonimo, respectivamente, una vez más, con la llegada del Renacimiento, autores como Huarte señalaron que para considerar que una persona contaba con ingenio y habilidad eran necesarias cuatro condiciones en el cerebro a difernecia de Descartes, Locke y Hobbes, los cuales tendrían una visión difrente con la mente, ya que estos últimos la considerarían de una forma científica y mecánica.

Entonces, ¿qué es la inteligencia?, ¿qué nos hace seres inteligentes? La inteligencia la definimos como aquella capacidad con la que se cuenta para resolver problemas, la cual está directamente relacionada con la capacidad intelectual, así como las habilidades para el razonamiento verbal, espacial y numérico. El psicólogo Raymond Cattell, propuso el concepto de inteligencia fluida, la cual es aquella en la que los individuos utilizan a su favor todos aquellos conocimientos que se han obtenido en el proceso de desarrollo para resolver problemas a los cuales nunca se habían enfrentado, esto a través del razonamiento, la capacidad de análisis y la habilidad para extrapolar conceptos que permitan abordar nuevos problemas de una forma innovadora, por lo que Raymond Cattell, considera que la inteligencia no es saber mucho, sino saber qué hacer con todo aquello que se conoce.  
Según la neuropediatra María José Mas, comenta que la inteligencia va relacionada con el neurodesarrollo el cual es un proceso que se realiza a través de la herencia genética de cada una de las personas donde el cerebro crece y genera la capacidad de desarrollar las habilidades de los seres humanos.

Sin embargo, estas definiciones se ven reflejadas constantemente con el desempeño que tienen las personas en el ámbito académico dejando de lado todo lo que tiene que ver con las hábilidades blandas, como lo son el manejo de als emocioens y las relaciones sociales, las cuales, juegan un papel muy importante en la resolución de muchos problemas a los que nos enfrentamos los seres humanos.

Debido a esta deficiencia en la definición de lo que hoy en día consideremos como inteligencia, es que psicólogos como Howard Gardner han propuesto nuevos planteamientos de la inteligencia, como es el caso de la Teoría de las Inteligencias Múltiples generado por Gardner, en la que se considera que los seres humanos contamos con competencias específicas en las que nos desarrollamos en múltiples áreas del razonamiento más alla del lógico-matemático, como lo son las inteligencias musical, corporal, espacial, linguistica, naturalista e interpersonal.

## Desarrollo
Como bien sabemos la teoría de las Inteligencias Múltiples fue ideada por el psicólogo estadounidense Howard Gardner con la finalidad de esclarecer lo que hasta ese entonces coonociamos como única inteligencia, para ello Gardner propuso que la vida humana requiere desarrollara varios tipo de inteligencia, ya que cada uno de estos tipos engloba una serie de habilidades que involucran la capacidad de pensamiento abstaracto más allá del conocimiento que convecionalemnente consideramos como inteligencia, es decir, las notas en el salón de clases, o bien, ser ágil mentalmente, ya que cada uno de estos tipo de inteligencias nos son útiles para afrontar los diferentes desafíos que se nos presentan día con día.

Gadner no buscaba contradecir la definición que tenemos actualmente de la inteligencia, es decir, "aquella capacidad de solucionar problemas o elaborar bienes valiosos", sino al contrario, busca replantear esta idea, cosa que adempas de hacerlo de una forma muy radical, también fue polémica en su momento, por lo que la Teoría de las Inteligencias Múltiples se vuelve conocida e interesante.

¿Existe más de un tipo de inteligencia? Gardner en su teoría nos plantea la idea de que efectivamente existe más allá de una única inteligencia, fue en el año de 1983, cuando publica su primer libro *Frames of Mind: The Thory of Multiple Intelligences*, en la que defiende completamente la idea de que la inteligencia debe ser comprenddido como varias habilidades con las que cuenta el ser humano para el precesamiento de la información y que funciona de manera paralela e independientes unas de otras, lo que significaría que si una persona cuenta con muy poco CI (cociente intelectual), esta misma puede ser demasiado inteligente en otros ámbitos los cuales no son tomados en cuenta por este tipo de herramientas de psicometría. Gardner tomaba inspiración en las teorías de construccionismo social para decir que lo que suele ser llamado "inteligencia" es un concepto que ha sido construido por la sociedad y que no tiene porqué estar exluido de otros factores como los son fenómenos históricos y culturaes que influeyen en la manera en la que medimos las facultades mentales y no ser solo un elemento psicológico natural.

### Teoría de las Inteligencias Múltiples
Howard Gardner en colaboración con la Universidad de Harvard, advirtieron que la inteligencia académica, es decir, obtener las mejores notas, titularse y obtener méritos educativos, no son un factor por el cual se puede conocer la inteligencia de una persona, por lo que era necesario contar con un concepto de "inteligencia", tal era el caso de estudiantes los cuales a pesar de obtener excelentes notas académicas, presentan problemas importantes en otros ámbitos como es en el caso de relacionarse con otras personas o mejorar facetas de su vida, con esto, Gardner propondría la idea y afirmaría que Stephen Hawking no posee una mayor inteligencia que Leo Messi, más cada una de estas personas ha desarrollado su inteligencia en un ámbito diferente.

#### Inteligencia linguística
Esta se denomina como aquella capacidad con la que cuentan los seres humanos para poderse comunicar entre ellos, lo cual es travservar a todas las culturas existentes. Esta no solo hace referencia a la capacidad de comunicación oral, sino a otras formas en las cuales nos comunicamos como es el caso de la escritura, gestualidad, entre varias otras.
